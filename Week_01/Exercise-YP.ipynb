## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100? 
df = pd.read_csv('2024 DS Wed listings.csv') ## stores read csv file in df
price_filter = df[df['price'] < 100] ## should only read prices that are less than 100.

count_listings = len(price_filter) ## reads the number of listings that have prices less than 100.
print(count_listings) ## this will print the number of listings with prices that are less than 100.

## There are 22778 listings with a price less than 100.


##print(price_filter) ##print the filter
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.

df_bk = pd.read_csv('2024 DS Wed listings.csv')
location_filter = df_bk[df_bk['neighbourhood_group'] == 'Brooklyn'] ##within purple parenthesis the value must have the name of column from the data set.

count_listings = len(location_filter)
print(count_listings)

## There are 18632 listings that are in brooklyn.
```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.
df = pd.read_csv('2024 DS Wed listings.csv')
price_location_filter = df[(df['price'] < 100) & (df['neighbourhood_group'] ==  'Brooklyn')]
count_listings = len(price_location_filter)
print(count_listings)

## There are 10473 listings that are in brooklyn and with a price less than 100.

```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
# How many total are there that have those names
df = pd.read_csv('2024 DS Wed listings.csv')
host_name_filter = df[df['host_name'].isin(['Michael', 'David', 'John', 'Daniel'])]
count_listings = len(host_name_filter)
print(count_listings)

## There are 1258 listings that have the host_names of 'Michael', 'David', 'John', 'Daniel'.

```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 

df = pd.read_csv('2024 DS Wed listings.csv') 

## how do I create a new column?
## df['column_name'] = np.where(df['column_x'], df['column_y'])
df['adjusted_price'] = np.where(df['neighbourhood_group'] == 'Williamsburg', df['price'] + 100, df['price']) ## this makes the adjusted price column that adds $100 to every williamsburg listing.

df.to_csv('updated_file.csv', index=False) ## writes the updated dataframe back into the csv file.

print(df) ##this prints the data frame with the current updates.

##pd.read_csv('2024 DS Wed listings.csv') 
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  
df = pd.read_csv('2024 DS Wed listings.csv')

total_rooms =len(df) ## calculates the total number of rooms in general ==> my denominator
private_count = df[df['room_type'] == 'Private room'].shape[0] ## creating my private count to find the number of private rooms. ==> numerator 1
shared_count = df[df['room_type'] == 'Entire home/apt'].shape[0] ## creating my shared count to find the number of shared rooms. ==> numerator 2

## divide the numerators over my denominator
private_rooms_percentage = (private_count / total_rooms) * 100 
shared_rooms_percentage = (shared_count / total_rooms) * 100

print(f"Private Rooms: {private_rooms_percentage:.2f}%") ## prints the percentage of my private rooms, only up to 2 decimal places.
print(f"Shared Rooms: {shared_rooms_percentage:.2f}%") ## prints the percentage of my shared rooms, only up to 2 decimal places.

##pd.read_csv('2024 DS Wed listings.csv')
##`.value_counts()` replace value with actual thing you are trying to count.

## I used 'Entire home/apt' for shared room because that was how shared rooms are stated in the CSV file. 
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.

## how to use groupby??? ==> df.groupby('column_name')
##pd.read_csv('2024 DS Wed listings.csv')

df = pd.read_csv('2024 DS Wed listings.csv')

neighbourhood_count = df.groupby('neighbourhood_group').size() ## groups dataframe by the given column then counts the number of rows in each group.

print(neighbourhood_count)
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 

##pd.read_csv('2024 DS Wed listings.csv')

df = pd.read_csv('2024 DS Wed listings.csv')

##np.mean?? ==> wasn't needed.
neighbourhood_mean_price = df.groupby('neighbourhood_group')['price'].mean()
print(neighbourhood_mean_price)
```

```python
# 2.5. Using `groupby`, find the mean price for each room_type. 
##pd.read_csv('2024 DS Wed listings.csv')

df = pd.read_csv('2024 DS Wed listings.csv')

room_type_mean_price = df.groupby('room_type')['price'].mean()
print(room_type_mean_price)
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 

df = pd.read_csv('2024 DS Wed listings.csv')

neighbourhood_groups = df.groupby('neighbourhood_group') ## groupby

result = neighbourhood_groups.agg({         # how to use .agg function and set min + max for each price
    'price' : ['min', 'max']
})


print(result)
```

```python
# 4. Using `groupby`, find the mean price for each room_type in each neighbourhood_group.

df = pd.read_csv('2024 DS Wed listings.csv')

ng_rt_grouped = df.groupby(['neighbourhood_group', 'room_type'])

result = ng_rt_grouped.agg({
        'price': ['mean']
    })

print(result)
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.

## basically inculde count, min, max and median in the agg.

df = pd.read_csv('2024 DS Wed listings.csv')

ng_rt_grouped = df.groupby(['neighbourhood_group', 'room_type'])

result = ng_rt_grouped.agg({
        'price': ['count','min','max','mean','median']
    })

print(result)
```
